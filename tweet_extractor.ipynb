```python
hashtag = "#ElDebateDefinitivo"
query = tw.Cursor(api.search, q=hashtag, lang="es", since="2021-05-24").items(2000)
tweets = [{'Tweet':tweet.text, 'Timestamp':tweet.created_at} for tweet in query]
print(tweets)
```

```python
df = pd.DataFrame.from_dict(tweets)
df.head(2000)
```

```python
petro_handle = ['GustavoPetro', 'Gustavo Petro', 'Gustavo', 'Petro', 'petrogustavo']
fico_handle = ['FICO', 'Fico Gutierrrez', 'Fico', 'Federico', 'Federico Gutierrez', 'FicoGutierrez']
rodolfo_handle = ['RodolfoHernandez', 'Rodolfo Hernandez', 'Rodolfo', 'Ingeniero Rodolfo', 'ingrodolfohdez']
# fajardo_handle = ['sergio_fajardo', 'Sergio Fajardo', 'Fajardo']
```

```python
def identify_subject(tweet, refs):
    flag = 0 
    for ref in refs:
        if tweet.find(ref) != -1:
            flag = 1
    return flag

df['Petro'] = df['Tweet'].apply(lambda x: identify_subject(x, petro_handle)) 
df['Fico'] = df['Tweet'].apply(lambda x: identify_subject(x, fico_handle))
df['Rodolfo'] = df['Tweet'].apply(lambda x: identify_subject(x, rodolfo_handle))
# df['Fajardo'] = df['Tweet'].apply(lambda x: identify_subject(x, fajardo_handle))

df.head(2000)
```

```python
df = df.drop_duplicates(subset=['Tweet'], keep='first')
df
```
